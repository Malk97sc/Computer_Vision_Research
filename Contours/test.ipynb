```python
import cv2 as cv 
import numpy as np
import matplotlib.pyplot as plt

#my functions
from FindContour import FindContour
from Mask import clean_image, segment_image
```

```python
path = "../Data/hand.jpg"
img = cv.imread(path, cv.IMREAD_GRAYSCALE)
img.shape
```

```python
def plot_img(img, cmp = True):
    if cmp == False:
        plt.imshow(img)    
    else: 
        plt.imshow(img, cmap = plt.cm.gray)
    plt.axis("off")
    plt.show()
```

```python
plot_img(img)
```

```python
thres = segment_image(img, 242)
thres = cv.bitwise_not(thres)
plot_img(thres)
```

```python
find = FindContour(thres)
contour = find.search_contour()
contourCV = find.contour2CV(contour)
```

```python
img_out = cv.cvtColor(img.copy(), cv.COLOR_GRAY2BGR)

cv.drawContours(img_out, contourCV, -1, (0, 255, 0), 3)
plot_img(img_out)
```
