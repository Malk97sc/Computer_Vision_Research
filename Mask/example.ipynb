# Example Using Segment

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from segment_image import segment_image
from clean_image import clean_image
```

## Read Data

```python
path = '../Data/rice.jpeg'
img = cv.imread(path, cv.IMREAD_GRAYSCALE)
img.shape
```

## Defining a Threshold

```python
thres = 125
thres_clean = 50
```

## Image Without Clean

```python
result = segment_image(img, thres)
```

```python
plt.imshow(result, cmap = plt.cm.Grays)
plt.show()
```

## Clean Image

```python
clean = clean_image(img)
clean_segm = segment_image(clean, thres_clean)
```

```python
plt.imshow(clean_segm, cmap = plt.cm.Grays)
plt.show()
```
