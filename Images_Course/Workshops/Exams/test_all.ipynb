```python
import pandas as pd
import numpy as np
import cv2 as cv
import os
import matplotlib.pyplot as plt

from Plot import plot_img, plot_cv
from Exam import ExamDetection
```

```python
path = "../../../Data/exams"
elements = os.listdir(path)
path_len = len(elements)
path_len
```

```python
for i in range(path_len):
    img_path = f"{path}/AI-HR0{i+1}.jpg"
    img_rgb = cv.imread(img_path, cv.IMREAD_COLOR_RGB)
    
    detector = ExamDetection(img_rgb)
    answers = detector.show_answers()
    img_result = detector.show_result(answers)

    fig, axes = plt.subplots(1, 2, figsize=(12, 6))
    axes[0].imshow(img_rgb)
    axes[0].set_title("Original")
    axes[0].axis("off")

    axes[1].imshow(img_result)
    axes[1].set_title("Detected Answers")
    axes[1].axis("off")

    plt.suptitle(f"Image {i+1}")
    plt.tight_layout(rect=[0, 0, 1, 0.95]) 
    plt.show()
```
