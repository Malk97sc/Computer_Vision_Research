# Calculator

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

#my functions
from ImageCalculator import ImageCalculator
from pplot import plot_cv, plot_img
```

```python
path = "../../../Data/hummingbird.jpg"
path2 = "../../../Data/tree.png"

img = cv.imread(path, cv.IMREAD_COLOR_RGB)
img2 = cv.imread(path2, cv.IMREAD_COLOR_RGB)
img.shape, img2.shape
```

## Resize

```python
size_img = 500
img = cv.resize(img, (size_img, size_img))
img2 = cv.resize(img2, (size_img, size_img))
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img)
plt.axis('off') 

plt.subplot(1, 2, 2)
plt.imshow(img2)
plt.axis('off') 

plt.tight_layout()
plt.show()
```

## Image Calculator Instance

```python
calc = ImageCalculator(img)
```

```python
sum = calc.sum_images(img2)
plot_img(sum)
```

```python
neg = calc.negative()
plot_img(neg)
```

```python
rot = calc.rotate(30)
plot_img(rot)
```

```python
sq = calc.square()
plot_img(sq)
```

```python
trans = calc.translate(20, 20)
plot_img(trans)
```
