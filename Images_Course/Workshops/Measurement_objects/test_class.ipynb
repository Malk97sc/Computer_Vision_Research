```python
def print_summary(results_df, method_used, pixel_cm):
    print(f"MEASUREMENT SUMMARY")
    print(f"Method: {method_used.upper()}")
    print(f"Conversion factor: {pixel_cm:.2f} px/cm")
    
    print("\nObjects Measurements:")
    print(results_df)
    
    print("STATISTICS:")
    print(f"Total objects measured: {len(results_df)}")
    print(f"Average dimensions: {results_df['Length_cm'].mean():.2f} x {results_df['Width_cm'].mean():.2f} cm")
    print(f"Largest object: {results_df['Length_cm'].max():.2f} x {results_df['Width_cm'].max():.2f} cm")
    print(f"Smallest object: {results_df['Length_cm'].min():.2f} x {results_df['Width_cm'].min():.2f} cm")
```

```python
method = "min_area_rect"
measurer = ObjectMeasurement(img_rgb, method = method)
```

```python
img_result, results, results_df = measurer.run_measurement()
```

```python
print_summary(results_df, method, measurer.pixel_cm)
```

```python
plot_img(img_result, size=(12, 16))
```
