# **Cargamos Librerías**

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

## **Leemos y Mostramos la Imagen**

```python
img = cv2.imread('onerice.jpeg')
plt.imshow(img)
plt.show()
```

## **Datos Imagen**

```python
def Datos(img):
  print('tamaño:',img.shape)
  print('maximo:',img.max())
  print('minimo:',img.min())
```

```python
Datos(img)
```

## **Extraemos un Canal**

```python
Xred   = img[:,:,0]
Xgreen = img[:,:,1]
Xblue  = img[:,:,2]
```

```python
Datos(Xred)
plt.imshow(Xred,cmap='gray')
plt.show()
```

## **Umbralizamos la Imagen**

```python
def segmentar(X,t):
  M,N = X.shape
  Y = np.zeros((M,N))
  Area = 0
  for i in range(M):
    for j in range(N):
      if X[i,j] > t:
        Y[i,j] = 255
        Area+=1
  print('Area',Area)
  return Y
```

```python
Y = segmentar(Xred,125)
plt.imshow(Y)
plt.show()
```

```python
img = cv2.imread('rices.png')
X   = img[:,:,0]
plt.imshow(X,cmap='gray')
plt.show()
```

```python
Y = segmentar(X,120)
plt.imshow(Y,cmap='gray')
plt.show()
```

```python
(M,N) = X.shape
xp = np.zeros((M,N),np.uint8)
for i in range(M):
  xmin = np.min(X[i,:])
  xp[i,:] = X[i,:] - xmin

plt.imshow(xp,cmap='gray')
plt.show()
```

```python
Y = segmentar(xp,50)
plt.imshow(Y,cmap='gray')
plt.show()
```
