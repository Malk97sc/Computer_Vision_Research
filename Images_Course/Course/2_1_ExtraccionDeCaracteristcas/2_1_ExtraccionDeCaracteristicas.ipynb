```python
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

## Harris Corner Detector

```python
img1 = cv2.imread('images/simple.jpg')
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img1 = cv2.resize(img1, (500, 500))

plt.figure(figsize=(10,10))
plt.imshow(img1)
plt.show()
```

```python
img1gray = cv2.cvtColor(img1, cv2.COLOR_RGB2GRAY)
img1gray = np.uint8(img1gray)

plt.figure(figsize=(8,8))
plt.imshow(img1gray, cmap='gray')
plt.show()
```

cv2.cornerHarris(img, blockSize, ksize, k)  

**Parametros de la función:**  
**img** - Imagen de entrada. Debe estar a escala de grises y ser de tipo float32.  
**blockSize** - Es el tamaño del vecindario a tener en cuenta por cada esquina.  
**ksize** - Parámetro de apertura de la derivada de Soble.  
**k** - Parámetro libre del algoritmo de detección de bordes de Harris.  

```python
dst = cv2.cornerHarris(img1gray, 2, 3, 0.04)
```

```python
dst = cv2.dilate(dst, None)
```

```python
img1res1 = img1.copy()
```

```python
img1res1[dst>0.01*dst.max()] = [0, 0, 255]
```

```python
plt.figure(figsize=(8,8))
plt.imshow(img1res1)
plt.show()
```

# Detector de Contorno

```python
img2 = cv2.imread('images/simple2.jpg')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img2 = cv2.resize(img2, (500, 500))

plt.figure(figsize=(8,8))
plt.imshow(img2)
plt.show()
```

```python
img2gray = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
img2gray = np.uint8(img2gray)

plt.figure(figsize=(8,8))
plt.imshow(img2gray, cmap='gray')
plt.show()
```

```python
ret, thresh = cv2.threshold(img2gray, 50, 255, 0)

plt.figure(figsize=(8,8))
plt.imshow(thresh, cmap='gray')
plt.show()
```

```python
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
img2res1 = cv2.drawContours(img2, contours, -1, (0,0,255), 3)
```

```python
plt.figure(figsize=(8,8))
plt.imshow(img2res1)
plt.show()
```
