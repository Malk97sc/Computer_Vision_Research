## Paquetes Necesarios

```python
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

## Leer Imagenes

Para leer una imagen se utiliza la libreria OpenCv (import cv2).  
En esta existe una función que permite la lectura.  
La imagen leida se guardará en un array tipo NumPy.

```python
img = cv2.imread('images/test.png')
```

Para visualizar la imagen se utiliza la biblioteca matplotlib.  
En ella se encuentra una función que permite su visualización.  

**Nota:** La biblioteca OpenCv lee la imagen con los espacios de colores invertidos, es decir, lee la imagen BGR en lugar de RGB. Por esto, es necesario realizar la conversión para poder visualizarlo correctamente.

```python
plt.imshow(img)
plt.show()
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
plt.show()
```

## Escalamiento de Imagenes

Para realizar un escalamiento de la imagen, OpenCV tiene una función que permite realizar dicho proceso.  
### **cv2.resize(src, dsize)**  
Es necesario definir los dos primeros parámetros:
 * El primero, la imagen a escalar
 * El segundo, una tupla donde se especifica la anchura y altura objetiva

```python
img1024 = cv2.resize(img, (1024, 1024))
```

```python
plt.imshow(img1024)
plt.show()
```

Se pueden utilizar diferentes métodos de interpolación para escalar una imagen. Este parámetro es opcional y puede influir muy fuerte en el resultado.

```python
img64_intercubic = cv2.resize(img, (64, 64), interpolation=cv2.INTER_CUBIC)
```

```python
plt.imshow(img64_intercubic)
plt.show()
```

```python
img64_interarea = cv2.resize(img, (64, 64), interpolation=cv2.INTER_AREA)
```

```python
plt.imshow(img64_interarea)
plt.show()
```

## Modelo de Colores

### RGB

```python
imgR = img[ : , : , 0]
imgG = img[ : , : , 1]
imgB = img[ : , : , 2]

fig, ax = plt.subplots(nrows=1, ncols=4, figsize=(15,5))

ax[0].imshow(img)
ax[1].imshow(imgR, cmap='Reds')
ax[2].imshow(imgG, cmap='Greens')
ax[3].imshow(imgB, cmap='Blues')

ax[0].set_title('Original')
ax[1].set_title('Canal Rojo - R')
ax[2].set_title('Canal Verde - G')
ax[3].set_title('Canal Azul - B')

plt.show()
```

### HSV

```python
imgHSV = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)

imgH = imgHSV[ : , : , 0]
imgS = imgHSV[ : , : , 1]
imgV = imgHSV[ : , : , 2]

fig, ax = plt.subplots(nrows=1, ncols=4, figsize=(15,5))

ax[0].imshow(imgHSV, cmap='gray')
ax[1].imshow(imgH, cmap='gray')
ax[2].imshow(imgS, cmap='gray')
ax[3].imshow(imgV, cmap='gray')

ax[0].set_title('Original')
ax[1].set_title('Canal Matriz - H')
ax[2].set_title('Canal Saturación - S')
ax[3].set_title('Canal Brillo - V')

plt.show()
```

### HLS

```python
imgHLS = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)

imgH = imgHLS[ : , : , 0]
imgL = imgHLS[ : , : , 1]
imgS = imgHLS[ : , : , 2]

fig, ax = plt.subplots(nrows=1, ncols=4, figsize=(15,5))

ax[0].imshow(imgHLS, cmap='gray')
ax[1].imshow(imgH, cmap='gray')
ax[2].imshow(imgL, cmap='gray')
ax[3].imshow(imgS, cmap='gray')

ax[0].set_title('Original')
ax[1].set_title('Canal Matriz - H')
ax[2].set_title('Canal Luminosidad - L')
ax[3].set_title('Canal Saturación - S')

plt.show()
```
