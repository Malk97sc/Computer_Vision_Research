```python
plot_img(img)
```

```python
def search_edge(img, n = 3, tol = 20):
    height, width = img.shape
    edge = np.zeros((height, width), dtype = np.uint8)
    
    for i in range(height - n):
        for j in range(width - n):
            dx = abs(int(img[i, j]) - int(img[i+n, j]))   
            dy = abs(int(img[i, j]) - int(img[i, j+n])) 

            '''
            dx = abs(img[i, j] - img[i+n, j])   
            dy = abs(img[i, j] - img[i, j+n])  
            '''
            
            if dx > tol or dy > tol:
                edge[i, j] = 255  
    
    return edge
```

```python
red = img[:, :, 0]
plot_img(red)
```

```python
blur = cv.GaussianBlur(red, (5,5) ,0)
```

```python
edge = search_edge(blur)
plot_img(edge)
```
